```python
import os
import pandas as pd
import geopandas as gpd
```

```python
input_folder = os.path.join('..', '..', '..', 'Ecuador', 'Data', 'Structured')
output_folder = os.path.join('..', 'input', 'Ecuador', 'Subnational')
```

# Load Canton

```python
canton = gpd.read_file(os.path.join(input_folder, "Admin", "canton_cities.geojson"))
print(canton['DPA_CANTON'].duplicated().sum())
canton = canton.drop_duplicates(subset=['DPA_CANTON']) #en attendant de mieux choisir
print(canton['DPA_CANTON'].duplicated().sum())
canton
```

# Add agri

```python
agri = gpd.read_file(os.path.join(input_folder, "Supply", "agri.geojson"))
print(agri['DPA_CANTON'].duplicated().sum())
canton['agri_prod'] = canton['DPA_CANTON'].map(agri.groupby('DPA_CANTON')['Agri_prod'].sum())
canton
```

# Add other sectors

```python
#sectors = pd.read_csv(os.path.join(input_folder, "Sectors", "sectors.csv"))
sectors = pd.read_csv(os.path.join(input_folder, "Sectors", "sectors_from_ntw_data.csv"),
                     dtype={'ecuador_sector': str})
#supply = gpd.read_file(os.path.join(input_folder, "Supply", "all_2015.geojson"))
supply = pd.read_csv(os.path.join(input_folder, "Supply", "sales_sector_canton_from_ntw_data.csv"),
                    dtype={"DPA_CANTON": str})
supply = supply.rename(columns=sectors.set_index('ecuador_sector')['trigram'].to_dict())
print(supply['DPA_CANTON'].duplicated().sum())
supply.head()
```

```python
supply.columns[1:]
```

```python
# to check mapping
pd.concat([
    pd.Series(supply.columns[1:]),
    pd.Series(supply.columns[1:].map(sectors.set_index('ecuador_sector')['trigram']))
], axis=1).to_csv('tmp.csv')
```

```python
# merge
print(canton['DPA_CANTON'].duplicated().sum())
print(supply['DPA_CANTON'].duplicated().sum())
canton_with_supply = pd.merge(
    canton,
    supply,
    on='DPA_CANTON',
    how='left'
)
```

```python
for sector, letter in sectors.set_index('trigram')['letter'].to_dict().items():
    canton['source_'+sector] = canton['DPA_CANTON'].map(supply.groupby('DPA_CANTON')['source_'+letter].sum())
```

# Pop

```python
canton_with_supply['population'] = 10000
```

# Export

```python
canton_with_supply.to_file(os.path.join(output_folder, "canton_economic_data_from_ntw.geojson"), driver="GeoJSON")
```
